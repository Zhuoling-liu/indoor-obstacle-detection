```python
from ultralytics import YOLO

# Load pretrained YOLOv8 model (nano version, suitable for robot deployment)
model = YOLO('yolov8n.pt')

# Train the model
results = model.train(
    data='Indoor-Obstacle-Detection-1/data.yaml',  # Dataset config file
    epochs=50,           # Number of training epochs
    imgsz=640,           # Image size
    batch=16,            # Batch size (reduce if GPU memory is insufficient)
    patience=10,         # Early stopping: stop if no improvement for 10 epochs
    device='mps',        # Use 'mps' for macOS, 'cuda' for NVIDIA GPU, otherwise 'cpu'
    project='obstacle_detection',  # Folder to save results
    name='yolov8n_indoor'          # Experiment name
)
```

```python
# Evaluate model performance
metrics = model.val()
print(f"mAP50: {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")
```

```python
# Test: Run inference on a single image
import glob

# Get test images
test_images = glob.glob('Indoor-Obstacle-Detection-1/test/images/*.jpg')

# Run prediction
results = model.predict(
    source=test_images[0],  # Test on the first image
    save=True,              # Save result images
    conf=0.5                # Confidence threshold
)

# Display detection results
from PIL import Image
import matplotlib.pyplot as plt

# Show original image and detection results
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(Image.open(test_images[0]))
axes[0].set_title('Original Image')
axes[0].axis('off')

# Detection result saved in runs/detect/predict/ folder
result_img = results[0].plot()  # Get annotated image
axes[1].imshow(result_img)
axes[1].set_title('Obstacle Detection Result')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```
